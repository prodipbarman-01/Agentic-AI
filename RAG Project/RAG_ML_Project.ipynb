```python
from langchain.document_loaders import PyPDFLoader

loader = PyPDFLoader("ml_notes.pdf.pdf")
documents = loader.load()

print("Total pages loaded:", len(documents))
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

texts = text_splitter.split_documents(documents)

print("Total chunks created:", len(texts))
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
from langchain.vectorstores import FAISS

vectorstore = FAISS.from_documents(texts, embedding_model)

print("Vector store created successfully!")
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k": 2})
```

```python
from langchain.llms import HuggingFaceHub
from langchain.chains import RetrievalQA

# যদি OpenAI না ব্যবহার করো, simple local LLM ব্যবহার করবো
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

llm = HuggingFacePipeline(pipeline=generator)

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever
)
```

```python
query = "What is machine learning?"
result = qa_chain.run(query)

print(result)
```

```python
query1 = "What is machine learning?"
print("Q1:", query1)
print("Answer:", qa_chain.run(query1))
print("\n")
```

```python
query2 = "What are the types of machine learning?"
print("Q2:", query2)
print("Answer:", qa_chain.run(query2))
print("\n")
```
